## Dot product example (nut butter sales)


```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
#number of jars sold
sales_amounts =  np.random.randint(20 , size= (5,3))
sales_amounts
```

```python
# creating weekly_sales dataframe

weekly_sales = pd.DataFrame(sales_amounts, index = ["Mon", "Tues" , "Wed", "Thurs", "Fri"],
                           columns =["Almond", "Peanut butter" , "Cashew butter"])
weekly_sales
```

```python
# creating prices array
prices = np.array([10,8,12])
prices
```

```python
#creating butter prices Dataframe
butter_prices = pd.DataFrame(prices, index = ["Price"],
                             columns = ["Almond", "Peanut butter" , "Cashew butter"])
butter_prices
```

```python
prices.shape
```

```python
#creating butter prices Dataframe
butter_prices = pd.DataFrame(prices.reshape(1,3),
                             index = ["Price"],
                             columns = ["Almond", "Peanut butter" , "Cashew butter"])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts)
```

```python
sales_amounts.shape
```

```python
prices.shape
```

```python
#since shapes aren't aligned, let's transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# create daily_sales
butter_prices.shape , weekly_sales.shape
```

```python
daily_sales= butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
```
