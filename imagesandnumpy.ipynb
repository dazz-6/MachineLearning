### if your img is in rgb then r=0, g=1, b=2 
basically, we are doing slicing here. 
[:,:,0] means take all pixels, but only their red intensity values
[:,:,1], only green channel

```python
plt.imshow(pic_red[:,:,1])  #0-255
```

```python
plt.imshow(pic_red[:,:,2])
```

0= no colour present,pure black
255 = colour present, white
and colour can be any of rgb
