```python
import numpy as np
```

```python
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
#create X (Features matrix)
X = heart_disease.drop("target",axis=1)

#create Y (labels)
y = heart_disease["target"]
```

```python
#choosing right model & hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)
#we'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train ,y_test = train_test_split(X,y, test_size=0.2)  #80 for training and 20 for testing
```

```python
clf.fit(X_train,y_train);
```

```python
X_train
```

```python
X_test
```

```python
# #make a prediction
# y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_train
```

```python
y_test
```

```python
# 4. evaluate the model
clf.score(X_train , y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix ,accuracy_score 
```

```python
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# 5. Improve a model
#trying diff amount of n_estimators

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on teest set: {clf.score(X_test,y_test) * 100:.2f}%")
```

```python
# 6. saving model and loading it
import pickle
pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
