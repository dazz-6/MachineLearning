## Arithmetic Operations on Images 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('input1.jpg')
```

```python
img2 = cv2.imread('input2.jpg')
```

```python
RGB_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
RGB_img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_img1)
```

```python
plt.imshow(RGB_img2)
```

### IMAGE ADDITION

```python
added = cv2.add(RGB_img1, RGB_img2)   #size must be equal
```

```python
resized_img2 = cv2.resize(RGB_img2 , (RGB_img1.shape[1] , RGB_img1.shape[0]))
```

```python
added = cv2.add(RGB_img1, resized_img2)
```

```python
plt.imshow(added)
```

### Weighted Addition 

```python
weighted =cv2.addWeighted(RGB_img1,0.7,resized_img2,0.3,0)
plt.imshow(weighted)
```

```python
weighted =cv2.addWeighted(RGB_img1,0.7,resized_img2,0.3,10)
plt.imshow(weighted)
```

### image substraction

```python
subtracted = cv2.subtract(RGB_img1 , resized_img2)
plt.imshow(subtracted)
```

```python
img3 = cv2.imread('random1.png')
RGB_img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
```

```python
img4 = cv2.imread('random2.png')
RGB_img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
```

```python
resized_img4 = cv2.resize(RGB_img4,(RGB_img3.shape[1] , RGB_img3.shape[0]))
```

```python
subtract = cv2.subtract(img3, resized_img4)
plt.imshow(subtract)
```

### Bitwise operator

```python
#Bitwise AND
and_img = cv2.bitwise_and(RGB_img3, resized_img4)
plt.imshow(and_img)
```

```python
#Bitwise OR
or_img = cv2.bitwise_or(RGB_img3 , resized_img4)
plt.imshow(or_img)
```

```python
#Bitwise XOR
xor_img = cv2.bitwise_xor(RGB_img3, resized_img4)
plt.imshow(xor_img)
```

```python
#Bitwise NOT
not_img = cv2.bitwise_not(RGB_img3 , resized_img4)
plt.imshow(not_img)
```
